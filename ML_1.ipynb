### KMeans clustering

```python
from sklearn.datasets import load_digits
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(69)
ds=load_digits()
arr=ds["data"]

def kmeans(k,arr,iter):
    n,d=arr.shape
    distances=np.zeros((n,k))
    centroids=arr[np.random.choice(n,k,replace=False)]
    cost=np.zeros(iter)
    for m in range(iter):
        for i in range(n):
            for j in range(k):
                distances[i][j]=np.linalg.norm(arr[i,:]-centroids[j])
        labels=np.argmin(distances,axis=1)
        for j in range(k):
            centroids[j]=np.mean(arr[labels==j],axis=0)
        cost[m]=np.sum(np.min(distances,axis=1))
    for i in range(len(centroids)):
        plt.subplot(2,5,i+1)
        plt.imshow(centroids[i].reshape(8,8),cmap=plt.cm.binary)
        plt.axis("off")


    plt.show()  
    plt.plot(cost,color="red")
    plt.xlabel("Iterations")
    plt.ylabel("cost")
    plt.show()
kmeans(10,arr,100)
```
