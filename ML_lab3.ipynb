```python
from sklearn.datasets import load_diabetes
x,y=load_diabetes(return_X_y=True)
import matplotlib.pyplot as plt
```

Linear Regression

```python
# Linear regression
from sklearn.linear_model import LinearRegression
modal_lin=LinearRegression()
modal_lin.fit(x,y)
# print(modal_lin.coef_)
# print(modal_lin.intercept_)
pred1=modal_lin.predict(x)
```

Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
modal_forest=RandomForestRegressor(random_state=1)
modal_forest.fit(x,y)
pred2= modal_forest.predict(x)
```

Gradient Boost Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
modal_grad=GradientBoostingRegressor(random_state=1)
modal_grad.fit(x,y)
pred3=modal_grad.predict(x)
```

1. Residual error 

```python
residual1=y-pred1
residual2=y-pred2
residual3=y-pred3
print("Linear:",sum(residual1))
print("Random_forest:",sum(residual2))
print("Gradient_boost:",sum(residual3))
```

2. prediction for first 20 rows

```python
print("Linear:",pred1[1:20])
print("Random_forest:",pred2[1:20])
print("Gradient_boost:",pred3[1:20])
```

3. plots


```python
plt.figure()
plt.plot(pred1[1:20],"gd",label='GradientBoostingRegressor')
plt.plot(pred2[1:20],"b^",label='RandomForestRegressor')
plt.plot(pred3[1:20],"ys",label='LinearRegression')

plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
plt.ylabel('Predicted')
plt.xlabel('Training samples')
plt.legend(loc="best")
plt.title('Comparision of different regressors')
plt.show()
```
