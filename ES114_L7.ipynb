### Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from mpl_toolkits import mplot3d

```

### Importing dataset

### Diabetes Dataset,  from N. Inst. of Diabetes & Diges. & Kidney Dis.
### link: https://www.kaggle.com/datasets/mathchi/diabetes-data-set

```python
diabetes = pd.read_csv('D:\Python\TermII\diabetes.csv')
```

```python
# coloumn names

data=diabetes[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].replace(0, np.NaN) 
data.fillna(data.mean(), inplace=True)
data.head()
```

### Making three clusters with labels 0, 1 and 2 which will show if there will be diabetes or the modal can't predict the result 

```python

%%timeit
kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(data)
# plt.hist(kmeans)
# plt.show()
```

### time in predicting labels using 8 colums is 
#### 83.6 ms ± 2.01 ms per loop (mean ± std. dev. of 7 runs, 10 loops each)

### Using PCA to reduce dimensions to 3 dimensions


```python
# ### Using PCA to reduce dimensions to 3 dimensions
pca = PCA(n_components=3)
pca.fit(data)
data3 = pca.transform(data)
```

### clustering 3 dimensional data using kmeans with k=3

```python

%%timeit
kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(data3)
```

### time in predicting labels using 3 colums is 
#### 81.2 ms ± 1.28 ms per loop (mean ± std. dev. of 7 runs, 10 loops each)

### Using PCA to reduce dimensions to 2 dimensions


```python
pca = PCA(n_components=2)
pca.fit(data)
data2 = pca.transform(data)
```

```python

%%timeit
kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(data2)
```

### time in predicting labels using 2 colums is 
#### 80.6 ms ± 1.38 ms per loop (mean ± std. dev. of 7 runs, 10 loops each)

# Making 2d and 3d plots to visualize the output

```python
# plot data2 and kmeans centers in 2D
kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(data2)
plt.scatter(data2[:,0], data2[:,1], c=kmeans)
# plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red')
```

```python
# plot data3 and kmeans centers in 3D

kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(data3)

ax = plt.axes(projection='3d')

ax.scatter3D(data3[:,0], data3[:,1],data3[:,2], c=kmeans)
plt.show()
```
