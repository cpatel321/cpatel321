Step 1: Simulation of a single trajectory of random walk on a 2-D grid:

```python
################ 2-D random walk model ###############################

import random
import math
import numpy as np
import matplotlib.pyplot as plt

# define the number of steps


def random_walk(n):
  dir=[[1,0],[0,1],[-1,0],[0,-1]]
  x_positions = [0]
  y_positions = [0]
  x = 0
  y = 0

# Start Random walk simulation for n steps:
# Loop over n steps and update the x and y coordinates:

  for i in range(n) :
    val = random.choice(dir) # choose a random number between 1 and 4 (uniform distribution) and move atom accordingly
    x=x+val[0]
    y=y+val[-1]      
    x_positions.append(x)
    y_positions.append(y)

  return x_positions,y_positions  
n=5000
x,y=random_walk(n)

### 

# Plot the trajectory:

plt.title("Random Walk ($n = " + str(n) + "$ steps)", fontsize = 28)
plt.plot(x,y)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("X", fontsize = 24)
plt.ylabel("Y", fontsize = 24)
plt.show()

# print the coordinate of the starting point
print(f"Initial position: ({x[0]}, {y[0]})")

# print the coordinate of the final point 
print(f"Final position after {n} steps: ({x[-1]}, {y[-1]})")
```

Step 2: Run the random walk simulation multiple times and average data to calculate the diffusion constant:

```python
######## Calculation of end_to_end distances, Mean squared displacement and the diffusion constant ##################

n_walks = 20  # number of independent random walk simulations
n = 5000     # number of steps in each random walk

msd_data = [] # to store the mean squared displacement data from each run
# print(msd_data.shape)
# Loop over n_walks number of independent simulations and accumulate the MSD data

for i in range(n_walks) :
    msd_s=[]
    def random_walk(n):
        global i
        global msd_s
        
        dir=[[1,0],[0,1],[-1,0],[0,-1]]
        x_positions = [0]
        y_positions = [0]
        x = 0
        y = 0

# Start Random walk simulation for n steps:
# Loop over n steps and update the x and y coordinates:

        for j in range(n) :
            val = random.choice(dir) # choose a random number between 1 and 4 (uniform distribution) and move atom accordingly
            x=x+val[0]
            y=y+val[-1]      
            x_positions.append(x)
            y_positions.append(y)
            dx =     x   # displacement in x direction, for each step
            dy =     y     # displacement in y direction, for each step
            msd=0
            msd = dx**2 + dy**2             # mean square displacement
            # print(msd)
            msd_s.append(msd)   # accumulate the mean square displacement values
        # return(msd_s)
    random_walk(n)
    msd_data.append(msd_s)

            # print(msd)
    # Enter code here
    # Make call to the random_walk function

    
    

#Average the MSD data over all the simulations:
# print(np.array(msd_data).shape)
msd_average = np.mean(msd_data, axis=0)                                           
# print(msd_average.shape)
# Plot the Mean squared displacement and extract the diffusion constant:
time = np.arange(n)


diffusion_coefficient = np.polyfit(time, msd_average, 1)[0] / 4.0

plt.figure(figsize=(10, 6))
plt.plot(time, msd_average, marker='o', color = 'cornflowerblue', linestyle='-', label='MSD')
plt.plot(time, 4 * diffusion_coefficient * time, linestyle='--',linewidth= 4,color = 'red', label='Fit: 4Dt')
plt.title("Mean squared Displacement (MSD)", fontsize = 28)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("No. of steps", fontsize = 24)
plt.ylabel("MSD", fontsize = 24)

print("Diffusion Coefficient:", diffusion_coefficient)
```
